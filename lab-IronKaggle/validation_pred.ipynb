```python
import pandas as pd
from sklearn.model_selection import train_test_split
import pickle
import numpy as np
```

```python
data = pd.read_pickle(r".\data\Validation_Clean.p")
```

```python
validation_features = data.drop(["True_index","State_holiday"], axis=1)
```

```python
gb_model = pickle.load(open(r".\models\gb_reg.p", "rb"))
```

```python
pred = gb_model.predict(validation_features)
```

```python
pred = pd.Series(pred)
```

```python
predicted_sales = pd.concat([data, pred], axis=1)
predicted_sales.head()
```

```python
predicted_sales[0] = np.where(predicted_sales["Open"] == 0, 0, predicted_sales[0])
predicted_sales.head()
```

```python
submit = predicted_sales[["True_index", 0]].rename(columns={0:"Predicted_Sales"})
submit.head()
```

```python
submit.to_csv(r".\PREDICTED.csv", index=False)
```
